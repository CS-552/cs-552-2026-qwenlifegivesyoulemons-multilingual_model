```python
### Testing the multiple datasets from exams
from datasets import get_dataset_config_names, load_dataset

DATASETS = [
    "haonan-li/cmmlu",
    "ceval/ceval-exam",
    "dvilares/head_qa",
    "mhardalov/exams",
    "manzar/kaleidoscope-bench-text-only-subset",
]
for path in DATASETS:
    print("=" * 70)
    print(path)
    try:
        cfgs = get_dataset_config_names(path)
        print("  configs:", cfgs)
        cfg = cfgs[0] if cfgs else None
        ds = load_dataset(path, cfg) if cfg else load_dataset(path)
        split = list(ds.keys())[0]
        print("  splits :", list(ds.keys()))
        print("  columns:", ds[split].column_names)
        print("  sample :", ds[split][0])
    except Exception as e:
        print("  FAILED:", repr(e))
```

```python
### testing the failed imports/missing reads
import json
from datasets import load_dataset
from huggingface_hub import list_repo_files, hf_hub_download

# 1. EXAMS
print("=" * 60, "\n[EXAMS multilingual]")
try:
    ds = load_dataset("mhardalov/exams", "multilingual")
    sp = list(ds.keys())[0]
    print("  splits:", list(ds.keys()), "| columns:", ds[sp].column_names)
    print("  sample:", ds[sp][0])
except Exception as e:
    print("  FAILED:", repr(e))

# 2. CMMLU
print("=" * 60, "\n[CMMLU raw files]")
try:
    files = list_repo_files("haonan-li/cmmlu", repo_type="dataset")
    csvs = [f for f in files if f.endswith(".csv")]
    print(f"  {len(csvs)} csv files; examples:", csvs[:6])
    p = hf_hub_download("haonan-li/cmmlu", csvs[0], repo_type="dataset")
    with open(p, encoding="utf-8") as f:
        for i, line in enumerate(f):
            print(f"  line {i}:", line.rstrip()[:180])
            if i >= 2:
                break
except Exception as e:
    print("  FAILED:", repr(e))

# 3. HEAD-QA
print("=" * 60, "\n[HEAD-QA raw files]")
try:
    files = list_repo_files("dvilares/head_qa", repo_type="dataset")
    print("  files:", files)
    jsons = [f for f in files if f.endswith(".json")]
    if jsons:
        p = hf_hub_download("dvilares/head_qa", jsons[0], repo_type="dataset")
        with open(p, encoding="utf-8") as f:
            data = json.load(f)
        if isinstance(data, dict):
            print("  top-level keys:", list(data.keys()))
        print("  preview:", json.dumps(data, ensure_ascii=False)[:700])
except Exception as e:
    print("  FAILED:", repr(e))
```

```python
import json, zipfile
from huggingface_hub import list_repo_files, hf_hub_download

# CMMLU — full file listing (no .csv found; what IS in there?)
print("=" * 60, "\n[CMMLU full file list]")
try:
    files = list_repo_files("haonan-li/cmmlu", repo_type="dataset")
    print("  all files:", files)
except Exception as e:
    print("  FAILED:", repr(e))

# HEAD-QA — look inside the zip
print("=" * 60, "\n[HEAD-QA zip contents]")
try:
    p = hf_hub_download("dvilares/head_qa", "data/head-qa-es-en-pdfs.zip",
                        repo_type="dataset")
    with zipfile.ZipFile(p) as z:
        names = z.namelist()
        print(f"  {len(names)} entries; first 20:", names[:20])
        jsons = [n for n in names if n.endswith(".json")]
        print("  json files:", jsons)
        if jsons:
            with z.open(jsons[0]) as f:
                data = json.load(f)
            if isinstance(data, dict):
                print("  json keys:", list(data.keys()))
            print("  preview:", json.dumps(data, ensure_ascii=False)[:800])
except Exception as e:
    print("  FAILED:", repr(e))
```
